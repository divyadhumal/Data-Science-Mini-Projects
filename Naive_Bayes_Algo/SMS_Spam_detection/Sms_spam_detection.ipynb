```python
import pandas as pd
```

```python
df= pd.read_csv("spam.csv", encoding='latin-1')
df = df[['v1','v2']]
df.columns = ['label','message']
```

```python
#convert label into numbers
df['label'] = df['label'].map({'ham':0,'spam':1})
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

# Step 1: split (text data)
X_train, X_test, y_train, y_test = train_test_split(
    df['message'], df['label'], test_size=0.2, random_state=42
)

# Step 2: vectorize ONLY ONCE
cv = CountVectorizer()
X_train = cv.fit_transform(X_train)   # text → numbers
X_test = cv.transform(X_test)

# Step 3: train
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
msg = ["Congratulations! You won a free ticket"]
msg=cv.transform(msg)
print(model.predict(msg))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print("Confusion Matrix")
print(confusion_matrix(y_test, y_pred))

print("Accuracy Score:", accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
